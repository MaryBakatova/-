**Стандартизация**

Цель стандартизации: Приведение всех признаков к одинаковому среднему значению и дисперсии. Этот метод часто используется в алгоритмах, чувствительных к масштабированию данных, таких как линейная регрессия, логистическая регрессия и SVM.

Подходы и алгоритмы

- **One-Hot Encoding** (Декодирование): Применяется для категориальных признаков. Каждое уникальное значение заменяется бинарной переменной.

- **Feature Scaling** (Масштабирование признаков): Включает методы нормализации и стандартизации.

- **Robust Scaler** (Робастное масштабирование): Использует медиану и межквартильный размах вместо среднего и стандартного отклонения, чтобы уменьшить влияние выбросов.

-** Power Transformer** (Преобразования степеней): Метод нелинейного преобразования данных, включая Box-Cox и Yeo-Johnson преобразования.

-** L1-Norm** (Манхэттенская норма): Эта методика нормализует вектор таким образом, чтобы сумма абсолютных значений элементов вектора равнялась единице.

-** L2-Norm** (Евклидова норма): Аналогично L1-норме, но здесь суммируются квадраты компонентов вектора

- **Softmax Normalization**: Этот метод чаще всего используется в задачах классификации, особенно в глубоком обучении. Он преобразует вектор в вероятности, такие что их сумма равна 1

- **Logistic Normalizatio**n: огистическое преобразование, которое ограничивает значения в диапазоне от 0 до 1

- **Binning** (Бинирование): Бинирование — это процесс группировки непрерывных данных в дискретные интервалы (бины). Например, возрастную группу можно разбить на бины: [18,25)[18,25), [25,35)[25,35), [35,45)[35,45) и т.д.

- **Outlier Detection** (Обнаружение выбросов): Выбросы могут значительно исказить результаты анализа. Для их обнаружения используют различные методики, такие как:

  - Метод межквартильного расстояния (IQR).
  - Метод Z-оценок.
  - Локально взвешенное сглаживание (LOESS).

- **Missing Value Imputation** (Заполнение пропущенных значений): Пропуски в данных могут сильно повлиять на результат анализа. Существуют различные стратегии заполнения пропусков:

  - Замена средним значением, медианой или модой.
  - Интерполяция (например, линейная интерполяция).
  - Использование моделей машинного обучения для предсказания пропущенных значений.

-** Dimensionality Reduction** (Снижение размерности): Когда данные содержат большое количество признаков, снижение размерности помогает упростить модель и избежать переобучения. Основные методы включают:

  - Principal Component Analysis (PCA).
  - Linear Discriminant Analysis (LDA).
  - t-SNE (для визуализации).

-** Discretization **(Дискретизация): Процесс преобразования непрерывных данных в дискретные категории. Часто используется для упрощения работы с данными в деревьями решений и байесовскими моделями.

-** Scaling by Maximum Absolute Value**: Масштабирование по максимальному абсолютному значению.

- **Winsorizing (Уинсоризация)**: Уинсоризация заключается в замене экстремальных значений на определённые пороговые значения (например, процентили), чтобы уменьшить влияние выбросов.
   
- **Variance Stabilizing Transformations **(Преобразования стабилизации дисперсии): В некоторых случаях дисперсия данных зависит от средних значений. Такие трансформации позволяют стабилизировать дисперсию. Примеры:
  - Анскомбовская трансформация.
  - Фишеровская z-трансформация.
    
- **Normalization by Deciles or Quantiles: **Нормализация данных путём приведения их к децилям или квантилю. Этот метод полезен, когда нужно учитывать распределение данных.
    
- **Decorrelation** (Декорреляция): Декорреляция уменьшает корреляцию между признаками, что полезно для многих алгоритмов машинного обучения. Один из подходов — использование PCA.
    
-** Kernel Density Estimation** (Оценка плотности ядра): Этот метод позволяет оценить плотность вероятностного распределения данных. Может использоваться для нормализации данных на основе оценки плотности.
    
- **Histogram Equalization** (Эквализация гистограммы): Методика выравнивания гистограмм изображений, который также применим к другим видам данных. Она улучшает контрастность и равномерность распределения значений.
    
-** Median Filtering** (Медианная фильтрация): Медианный фильтр удаляет шум из данных, заменяя каждое значение на медиану соседних значений. Особенно эффективен против импульсного шума.
    
- **Statistical Outlier Removal** (Статистическое удаление выбросов): Удаление выбросов на основе статистического анализа, например, через сравнение значений с медианой и межквартильным размахом.
    
- ***Polynomial Fitting*** (Полиномиальная аппроксимация): Полиномиальные функции используются для приближенного описания данных. Можно использовать для нормализации временных рядов или других типов последовательностей.

- **Wavelet Transformation** (Вейвлет-преобразование): Вейвлеты применяются для декомпозиции сигналов и удаления шумов. Могут быть полезны для нормализации сложных временных рядов.
    
- **Factor Analysis** (Факторный анализ): Метод снижения размерности, основанный на поиске скрытых факторов, объясняющих корреляции между признаками.
    
- **Whitening **(Отбеливание): Отбеливание — это процедура, при которой ковариационная матрица данных приводится к единичной матрице. Это делает признаки ортогональными и имеющими единичную дисперсию.
    
- **Data Balancing** (Балансировка данных): Балансировка классов полезна в задачах классификации, когда классы сильно несбалансированы. Используются техники оверсэмплинга и андерсэмплинга.
    
- **Shuffling** (Перемешивание): Перемешивание данных перед тренировкой модели помогает избежать смещения и улучшить обобщаемость модели.

# **5. Нормализация**
L1-Norm

```python
import numpy as np

def l1_normalize(vector):
    # Суммируем абсолютные значения всех элементов вектора
    norm = np.sum(np.abs(vector))

    if norm == 0:
        return vector
    else:
        return vector / norm

# Пример вектора
vector = np.array()[1, 2, 3]

# Нормализуем вектор
l1_norm_vector = l1_normalize(vector)
print(l1_norm_vector)
```

6. L2-Norm

```python
import numpy as np

def l2_normalize(vector):
    # Вычисляем евклидову норму вектора
    norm = np.linalg.norm(vector)

    if norm == 0:
        return vector
    else:
        return vector / norm

# Пример вектора
vector = [np.array(](np.array()[1, 2, 3])

# Нормализуем вектор
l2_norm_vector = l2_normalize(vector)
print(l2_norm_vector)
```

7. Softmax Normalization

```python
import numpy as np

def softmax(x):
    # Преобразуем входной массив в экспоненты
    exp_x = [np.exp(x](np.exp(x) - [np.max(x))](np.max(x)))

    # Нормируем результат
    return exp_x / [np.sum(exp_x)](np.sum(exp_x))

# Пример вектора
vector = (np.array()[1, 2, 3])

# Применяем функцию softmax
softmax_vector = softmax(vector)
print(softmax_vector)
```

8. Logistic Normalization

```python
import numpy as np

def logistic_normalize(x):
    # Логистическая функция
    return 1 / (1 + [np.exp(-x))](np.exp(-x)))

# Пример вектора
vector = [np.array(](np.array()[-1, 0, 1])

# Применяем логистическую нормализацию
logistic_vector = logistic_normalize(vector)
print(logistic_vector)
```

11. Декорреляция (PCA)

```python
import numpy as np
from sklearn.decomposition import PCA

# Генерируем коррелированные данные
np.random.seed(42)
n_samples = 1000
X = np.zeros((n_samples, 3))
X[:, 0] = np.random.normal(0, 1, n_samples)
X[:, 1] = X[:, 0] * 0.8 + np.random.normal(0, 0.3, n_samples)
X[:, 2] = X[:, 0] * 0.7 + np.random.normal(0, 0.4, n_samples)

def decorrelate_data(X):
    """
    Декорреляция данных с помощью PCA
    """
    pca = PCA(n_components=3)
    X_decorrelated = pca.fit_transform(X)

    return X_decorrelated, pca.explained_variance_ratio_

# Применяем декорреляцию
X_decorrelated, explained_variance = decorrelate_data(X)

print("Корреляционная матрица исходных данных:")
print(np.round(np.corrcoef(X.T), 2))

print("\nКорреляционная матрица после декорреляции:")
print(np.round(np.corrcoef(X_decorrelated.T), 2))

print("\nДоля объясненной дисперсии для каждого компонента:")
for i, var in enumerate(explained_variance, 1):
    print(f"Компонент {i}: {var:.2%}")
```

 13. Оценка плотности ядра (KDE)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde

# Генерируем данные с двумя модами
np.random.seed(42)
data = np.concatenate([
    np.random.normal(-2, 0.5, 500),
    np.random.normal(2, 0.5, 500)
])

def estimate_density(data, bandwidth=None):
    """
    Оценка плотности распределения данных
    """
    kde = gaussian_kde(data, bw_method=bandwidth)
    x_range = np.linspace(min(data), max(data), 200)
    density = kde(x_range)

    return x_range, density

# Оцениваем плотность
x_vals, density = estimate_density(data)

# Создаем график
plt.figure(figsize=(10, 6))
plt.plot(x_vals, density, 'b-', label='Оцененная плотность')
plt.hist(data, bins=30, density=True, alpha=0.5, color='gray',
         label='Гистограмма данных')
plt.title('Оценка плотности распределения')
plt.xlabel('Значение')
plt.ylabel('Плотность')
plt.legend()
plt.grid(True)
plt.savefig('kde_plot.png')

print("Основные характеристики распределения:")
print(f"Среднее: {np.mean(data):.2f}")
print(f"Медиана: {np.median(data):.2f}")
print(f"Стандартное отклонение: {np.std(data):.2f}")
```

14. Эквализация гистограммы

```python
import numpy as np
import matplotlib.pyplot as plt

# Генерируем данные с неравномерным распределением
np.random.seed(42)
data = np.concatenate([
    np.random.normal(-2, 0.5, 1000),
    np.random.exponential(1, 500),
    np.random.gamma(2, 1, 500)
])

def histogram_equalization(data):
    """
    Эквализация гистограммы данных
    """
    sorted_data = np.sort(data)
    cumulative_dist = np.arange(len(data)) / float(len(data))
    equalized_data = np.interp(data, sorted_data, cumulative_dist)

    return equalized_data

# Применяем эквализацию
equalized_data = histogram_equalization(data)

# Создаем графики
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

# Исходные данные
ax1.hist(data, bins=50, density=True, alpha=0.7, color='blue',
         label='Исходные данные')
ax1.set_title('Исходная гистограмма')
ax1.set_xlabel('Значение')
ax1.set_ylabel('Частота')
ax1.grid(True)
ax1.legend()

# Эквализированные данные
ax2.hist(equalized_data, bins=50, density=True, alpha=0.7, color='green',
         label='Эквализированные данные')
ax2.set_title('Гистограмма после эквализации')
ax2.set_xlabel('Значение')
ax2.set_ylabel('Частота')
ax2.grid(True)
ax2.legend()

plt.tight_layout()
plt.savefig('histogram_equalization.png')

print("Статистики исходных данных:")
print(f"Среднее: {np.mean(data):.2f}")
print(f"Медиана: {np.median(data):.2f}")
print(f"Стандартное отклонение: {np.std(data):.2f}")

print("\nСтатистики после эквализации:")
print(f"Среднее: {np.mean(equalized_data):.2f}")
print(f"Медиана: {np.median(equalized_data):.2f}")
print(f"Стандартное отклонение: {np.std(equalized_data):.2f}")
```

15. Медианная фильтрация

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import median_filter as mf

# Генерируем сигнал с шумом
np.random.seed(42)
t = np.linspace(0, 10, 1000)
signal = np.sin(t) + 0.5 * np.cos(3*t)
noise = np.random.normal(0, 0.1, len(t))
noisy_signal = signal + noise

# Добавляем импульсный шум
impulse_noise = np.random.choice([0, 2], size=len(t), p=[0.95, 0.05])
noisy_signal += impulse_noise

def median_filter(data, window_size=11):
    """
    Медианная фильтрация данных
    """
    filtered_data = mf(data, size=window_size)
    return filtered_data

# Применяем фильтрацию
filtered_signal = median_filter(noisy_signal)

# Создаем графики
plt.figure(figsize=(15, 5))

plt.subplot(131)
plt.plot(t[:200], noisy_signal[:200], 'r-', label='Шумной сигнал')
plt.title('Исходный сигнал со шумом')
plt.xlabel('Время')
plt.ylabel('Амплитуда')
plt.grid(True)
plt.legend()

plt.subplot(132)
plt.plot(t[:200], filtered_signal[:200], 'g-', label='Отфильтрованный сигнал')
plt.title('Сигнал после медианной фильтрации')
plt.xlabel('Время')
plt.ylabel('Амплитуда')
plt.grid(True)
plt.legend()

plt.subplot(133)
plt.plot(t[:200], signal[:200], 'b-', label='Идеальный сигнал')
plt.title('Идеальный сигнал')
plt.xlabel('Время')
plt.ylabel('Амплитуда')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.savefig('median_filtering.png')

print("Метрики качества сигнала:")
print(f"Средняя абсолютная ошибка до фильтрации: {np.mean(np.abs(noisy_signal - signal)):.3f}")
print(f"Средняя абсолютная ошибка после фильтрации: {np.mean(np.abs(filtered_signal - signal)):.3f}")
```

4. Декомпозиция данных

Декомпозиция данных — это процесс разделения данных на составляющие части для облегчения анализа, визуализации или дальнейшего моделирования. Существуют различные методы декомпозиции данных, каждый из которых имеет свои особенности и области применения.

4.1. Principal Component Analysis (PCA)

данных

Определение: PCA — это статистический метод, предназначенный для преобразования набора исходных переменных (признаков) в меньшее количество новых, ортогональных друг другу переменных, называемых главными компонентами. Эти новые переменные сохраняют максимальное количество информации, содержащейся в исходных данных.
Как работает PCA?

- Стандартизация данных: Для начала данные нормализуются, чтобы устранить влияние масштаба отдельных признаков.
- Поиск главных компонентов: Алгоритм находит такие направления в пространстве данных, вдоль которых наблюдается максимальная дисперсия (вариация). Эти направления называются главными компонентами.
- Выбор значимых компонент: В результате выбираются несколько первых главных компонент, которые объясняют большую часть общей дисперсии данных. Оставшиеся компоненты отбрасываются, что ведет к уменьшению размерности.

Преимущества PCA:

- Эффективное уменьшение размерности: Позволяет сократить число переменных до наиболее информативных, что упрощает дальнейший анализ и уменьшает вычислительные затраты.
- Простота интерпретации: Главные компоненты представляют собой линейные комбинации исходных признаков, что позволяет легко интерпретировать результаты анализа.

Недостатки PCA:

- Линейность: Метод эффективен только для линейных зависимостей между признаками. Если структура данных является нелинейной, PCA может оказаться недостаточно эффективным.

Применение PCA:

- Предобработка данных перед машинным обучением: Уменьшение размерности помогает избежать переобучения моделей и улучшить производительность алгоритмов.
- Анализ данных: Помогает визуализировать многомерные данные в двумерном или трехмерном пространстве.


```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Генерация тестовых данных (например, случайные точки в 3D-пространстве)
X = np.random.randn(1000, 3)

# Создание объекта PCA и подгонка модели к данным
pca = PCA(n_components=2)
pca.fit(X)

# Преобразование данных в пространство с двумя компонентами
X_reduced = pca.transform(X)

# Визуализация преобразованных данных
plt.figure(figsize=(8,6))
plt.scatter(X_reduced[:, 0], X_reduced[:, 1])
plt.xlabel('Первая главная компонента')
plt.ylabel('Вторая главная компонента')
plt.title('Результат применения PCA')
plt.show()
```

4.2. Independent Component Analysis (ICA)

ICA разделяет сигналы на независимые компоненты, предполагая, что они смешиваются линейно и являются негауссовыми. Этот метод используется для решения задач слепого разделения сигналов.

Преимущества:

- Хороший выбор для нелинейных данных.
- Эффективен для задач, связанных с обработкой сигналов.

Недостатки:

- Предполагает негауссовость компонентов.
- Может быть сложно интерпретировать результаты.

```python
import numpy as np
from sklearn.decomposition import FastICA
import matplotlib.pyplot as plt

# Генерация тестовых данных (смешивание сигналов)
np.random.seed(42)

# Исходные сигналы
signal_1 = np.sin(np.linspace(0, 10 * np.pi, 500))  # Синусоидальный сигнал
signal_2 = np.sign(np.cos(np.linspace(0, 20 * np.pi, 500)))  # Прямоугольный сигнал

# Смешанные сигналы
S = np.c_[signal_1, signal_2]
A = np.array([[1, 1], [0.5, 2]])  # Матрица смешивания
X = np.dot(S, A.T)  # Получение смешанных сигналов

# Применение ICA для восстановления исходных сигналов
ica = FastICA(n_components=2)
S_ = ica.fit_transform(X)  # Разделение сигналов
A_ = ica.mixing_.T  # Оцененная матрица смешивания

# Построение графиков
plt.figure(figsize=(12, 8))

plt.subplot(311)
plt.plot(signal_1, label='Источник 1')
plt.plot(signal_2, label='Источник 2')
plt.legend(loc='best')
plt.title('Исходные сигналы')

plt.subplot(312)
plt.plot(X[:, 0], label='Смесь 1')
plt.plot(X[:, 1], label='Смесь 2')
plt.legend(loc='best')
plt.title('Смешанные сигналы')

plt.subplot(313)
plt.plot(S_[:, 0], label='Разделенный сигнал 1')
plt.plot(S_[:, 1], label='Разделенный сигнал 2')
plt.legend(loc='best')
plt.title('Разделенные сигналы (ICA)')

plt.tight_layout()
plt.show()
```

Singular Value Decomposition (SVD)

SVD разлагает матрицу на три компоненты: ортогональную левую матрицу, диагональную матрицу сингулярных значений и ортогональную правую матрицу. SVD находит лучшие низкоразмерные представления данных.

Преимущества:

- Мощный инструмент для компрессии данных.
- Широко используется в рекомендательных системах и анализе текста.

Недостатки:

- Может быть трудно интерпретировать геометрически.

```python
import numpy as np

# Создаем тестовую матрицу
A = np.array([[3, 4, 3],
              [1, 2, 3],
              [4, 2, 1]])

# Выполняем SVD декомпозицию
U, s, Vt = np.linalg.svd(A)

# Выводим результаты
print("Исходная матрица A:")
print(A)
print("\nЛевая ортогональная матрица U:")
print(np.round(U, decimals=4))  # Округляем до 4 знаков после запятой
print("\nСингулярные значения s:")
print(np.round(s, decimals=4))
print("\nПравая ортогональная транспонированная матрица V^T:")
print(np.round(Vt, decimals=4))

# Проверяем свойства SVD
print("\nПроверка свойств SVD:")

# 1. Ортогональность U
print("\nU × U^T (должна быть единичная матрица):")
print(np.round(U @ U.T, decimals=4))

# 2. Ортогональность V
print("\nV × V^T (должна быть единичная матрица):")
print(np.round(Vt.T @ Vt, decimals=4))

# 3. Восстановление исходной матрицы
S = np.diag(s)
A_reconstructed = U @ S @ Vt
print("\nВосстановленная матрица A:")
print(np.round(A_reconstructed, decimals=4))
```

Non-negative Matrix Factorization (NMF)

NMF разлагает неотрицательные матрицы на две неотрицательные матрицы. Этот метод особенно полезен для задач, где данные имеют физическую интерпретацию, такую как интенсивность света или плотность населения.

Преимущества:

- Легко интерпретируется.
- Подходит для неотрицательных данных.

Недостатки:

- Ограничено неотрицательными данными.
- Может быть медленно вычисляться.

```python
import numpy as np

class NMF:
    def __init__(self, n_components, max_iter=100000, tol=1e-6):
        self.n_components = n_components
        self.max_iter = max_iter
        self.tol = tol

    def fit(self, X):
        """
        Разложение матрицы X на произведение двух неотрицательных матриц
        """
        K, N = X.shape
        W = np.random.rand(K, self.n_components)
        H = np.random.rand(self.n_components, N)

        for _ in range(self.max_iter):
            # Обновление H
            numerator_H = W.T @ X
            denominator_H = W.T @ W @ H + np.finfo(float).eps
            H = H * (numerator_H / denominator_H)

            # Обновление W
            numerator_W = X @ H.T
            denominator_W = W @ H @ H.T + np.finfo(float).eps
            W = W * (numerator_W / denominator_W)

        self.components_ = H
        self.transform_matrix_ = W
        return self

    def reconstruct(self):
        """
        Восстановление исходной матрицы из факторов
        """
        return self.transform_matrix_ @ self.components_

# Пример использования
np.random.seed(42)

# Создаём тестовый набор данных
X = np.array([
    [0, 1, 2, 3, 4],
    [0, 1, 2, 3, 2],
    [0, 0, 0, 0, 0],
    [7, 1, 1, 1, 1]
])

print("Исходная матрица:")
print(X)

# Применяем NMF с рангом 2
nmf = NMF(n_components=2)
nmf.fit(X)

print("\nМатрица W:")
print(nmf.transform_matrix_)
print("\nМатрица H:")
print(nmf.components_)

# Восстанавливаем матрицу
X_reconstructed = nmf.reconstruct()
print("\nВосстановленная матрица:")
print(X_reconstructed)

# Вычисляем ошибку
error = np.linalg.norm(X - X_reconstructed)
print(f"\nОшибка восстановления: {error:.8f}")
```

Canonical Correlation Analysis (CCA)

CCA находит максимальные корреляции между двумя множествами переменных. Этот метод используется для изучения связей между двумя наборами данных.

Преимущества:

- Полезен для мультидисциплинарных исследований.
- Хорошо работает с большими наборами данных.

Недостатки:

- Требует наличия двух наборов данных.
- Сложно интерпретировать результаты.

```python
import numpy as np
from sklearn.cross_decomposition import CCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Создаем два связанных набора данных
np.random.seed(42)

# Первый набор данных (X)
X = np.random.normal(0, 1, (100, 2))

# Второй набор данных (Y) создается как линейная комбинация X + шум
Y = 0.99999 * X + np.random.normal(0, 0.5, (100, 2))

# Масштабируем данные
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
Y_scaled = scaler.fit_transform(Y)

# Применяем CCA
cca = CCA(n_components=1)
X_cca, Y_cca = cca.fit_transform(X_scaled, Y_scaled)

# Оцениваем силу связи между наборами данных
score = cca.score(X_scaled, Y_scaled)
print(f'Коэффициент канонической корреляции: {score:.3f}')

# Визуализация результатов
plt.figure(figsize=(10, 5))

# График исходных данных
plt.subplot(121)
plt.scatter(X[:, 0], X[:, 1], label='X', alpha=0.5)
plt.scatter(Y[:, 0], Y[:, 1], label='Y', alpha=0.5)
plt.xlabel('Первая переменная')
plt.ylabel('Вторая переменная')
plt.title('Исходные данные')
plt.legend()

# График канонических компонент
plt.subplot(122)
plt.scatter(X_cca, Y_cca, alpha=0.5)
plt.xlabel('Каноническая компонента X')
plt.ylabel('Каноническая компонента Y')
plt.title('Канонические компоненты')

plt.tight_layout()
plt.savefig('cca_visualization.png')
plt.close()
```

Latent Dirichlet Allocation (LDA)

LDA — это метод тематического моделирования, который рассматривает документ как смесь различных тем, а темы — как смеси слов. LDA используется для анализа текста и обнаружения скрытых тем в документах.

Преимущества:

- Подходит для текстового анализа.
- Позволяет обнаруживать скрытые темы.

Недостатки:

- Требует большого объема данных.
- Требует настройки гиперпараметров.

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import gensim
from gensim.utils import simple_preprocess
from gensim.models import LdaModel
from gensim.corpora.dictionary import Dictionary

# Загрузка данных и предварительная обработка
def preprocess_text(text):
    """
    Функция для предварительной обработки текста:
    - Приведение к нижнему регистру
    - Удаление пунктуации
    - Токенизация
    - Удаление стоп-слов
    """
    # Приводим к нижнему регистру и удаляем пунктуацию
    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)

    # Токенизация
    tokens = word_tokenize(text)

    # Загружаем стоп-слова
    stop_words = set(stopwords.words('russian'))

    # Оставляем только значимые слова
    filtered_tokens = [token for token in tokens
                      if token not in stop_words]

    return filtered_tokens

# Создание корпуса документов
def create_corpus(documents):
    """
    Создает корпус документов из списка текстов
    """
    # Предобработка всех документов
    processed_docs = [preprocess_text(doc) for doc in documents]

    # Создаем словарь
    dictionary = Dictionary(processed_docs)

    # Создаем корпус в формате bag-of-words
    corpus = [dictionary.doc2bow(doc) for doc in processed_docs]

    return dictionary, corpus

# Обучение модели LDA
def train_lda_model(corpus, dictionary, num_topics=4):
    """
    Обучает модель LDA на подготовленном корпусе
    """
    lda_model = LdaModel(
        corpus=corpus,
        id2word=dictionary,
        passes=15,
        num_topics=num_topics,
        random_state=42
    )
    return lda_model

# Анализ документа
def analyze_document(document, lda_model, dictionary):
    """
    Анализирует отдельный документ и определяет его темы
    """
    # Предобработка документа
    bow = dictionary.doc2bow(preprocess_text(document))

    # Получаем распределение тем
    topics = lda_model.get_document_topics(bow)

    return topics

# Пример использования
if __name__ == "__main__":
    # Пример документов
    documents = [
        "Машинное обучение - это область искусственного интеллекта, которая занимается разработкой алгоритмов обучения на данных.",
        "Искусственный интеллект используется во многих областях современной жизни, от распознавания изображений до принятия решений.",
        "Нейронные сети являются ключевым инструментом глубокого обучения и применяются для решения сложных задач.",
        "Обработка естественного языка позволяет компьютерам понимать и генерировать человеческий язык."
    ]

    # Создаем корпус
    dictionary, corpus = create_corpus(documents)

    # Обучаем модель
    lda_model = train_lda_model(corpus, dictionary, num_topics=2)

    # Выводим темы
    print("Определенные темы:")
    topics = lda_model.print_topics(num_words=4)
    for topic_id, topic in topics:
        print(f"Тема {topic_id}:")
        print(topic)

    # Анализируем новый документ
    new_doc = "Нейронные сети используются для анализа больших объемов данных и прогнозирования результатов."
    topics = analyze_document(new_doc, lda_model, dictionary)
    print("\nАнализ нового документа:")
    for topic_id, prob in topics:
        print(f"Тема {topic_id}: {prob:.2%}")
```

Self-Organizing Maps (SOM)

SOM — это разновидность нейронных сетей, которые используют конкурентное обучение для организации данных в двумерное пространство. SOMs полезны для визуализации и кластеризации данных.

Преимущества:

- Визуально интуитивно понятен.
- Поддерживает нелинейные данные.

Недостатки:

- Сложность настройки топологии карты.
- Зависимость от начальных условий.



```python
import numpy as np
import matplotlib.pyplot as plt

class SOM:
    def __init__(self, input_size, map_size):
        """Инициализация SOM"""
        self.input_size = input_size
        self.map_size = map_size
        self.weights = np.random.rand(map_size[0], map_size[1], input_size)

    def find_bmu(self, input_vector):
        """Поиск лучшего соответствующего единства (BMU)"""
        distances = np.linalg.norm(self.weights - input_vector, axis=2)
        bmu_idx = np.unravel_index(np.argmin(distances), distances.shape)
        return bmu_idx

    def update_weights(self, input_vector, bmu_idx, learning_rate, neighborhood_radius):
        """Обновление весов"""
        for i in range(self.map_size[0]):
            for j in range(self.map_size[1]):
                dist = np.sqrt((i - bmu_idx[0])**2 + (j - bmu_idx[1])**2)
                if dist <= neighborhood_radius:
                    influence = np.exp(-(dist**2)/(2*neighborhood_radius**2))
                    self.weights[i,j] += learning_rate * influence * (input_vector - self.weights[i,j])

    def train(self, input_data, num_iterations=50, initial_lr=0.1, initial_radius=2):
        """Обучение SOM с уменьшенным количеством итераций"""
        for iter in range(num_iterations):
            lr = initial_lr * (1 - iter/num_iterations)
            radius = initial_radius * (1 - iter/num_iterations)

            for input_vec in input_data:
                bmu_idx = self.find_bmu(input_vec)
                self.update_weights(input_vec, bmu_idx, lr, radius)

# Создание тестовых данных
np.random.seed(42)
test_data = np.random.rand(20, 3)  # Уменьшили размер набора данных

# Создание и обучение SOM
som = SOM(input_size=3, map_size=(5, 5))  # Уменьшили размер карты
som.train(test_data)

# Визуализация результатов
plt.figure(figsize=(8, 8))
plt.imshow(som.weights.mean(axis=2), cmap='viridis')
plt.colorbar(label='Усредненное значение весов')
plt.title('Обученная SOM')
plt.savefig('som_result.png')  # Сохраняем график вместо показа
plt.close()
```

Kernel PCA

Kernel PCA — это расширение PCA, которое использует ядро для нахождения нелинейных главных компонент. Этот метод полезен для анализа данных с нелинейными структурами.

Преимущества:

- Обрабатывает нелинейные данные.
- Улучшенная способность к обобщению.

Недостатки:

- Вычислительно затратен.
- Чувствителен к выбору ядра.

```python
import numpy as np
from sklearn.datasets import make_moons
from sklearn.decomposition import KernelPCA
import matplotlib.pyplot as plt

# Создаем тестовый набор данных
X, y = make_moons(n_samples=200, noise=0.05, random_state=42)

# Создаем несколько визуализаций с разными значениями gamma
gammas = [1, 15, 30]
fig, axes = plt.subplots(1, len(gammas), figsize=(15, 5))

for idx, gamma in enumerate(gammas):
    kpca = KernelPCA(kernel='rbf', gamma=gamma, n_components=2)
    X_kpca = kpca.fit_transform(X)

    axes[idx].scatter(X_kpca[y==0, 0], X_kpca[y==0, 1], c='red', label='Класс 0')
    axes[idx].scatter(X_kpca[y==1, 0], X_kpca[y==1, 1], c='blue', label='Класс 1')
    axes[idx].set_title(f'Kernel PCA (gamma={gamma})')
    axes[idx].legend()

plt.tight_layout()
plt.savefig('kernel_pca_gamma.png')
```

Locally Linear Embedding (LLE)

LLE — это метод нелинейного уменьшения размерности, который пытается сохранить локальные свойства данных. LLE находит низкоразмерные представления, которые сохраняют соседство точек в исходном пространстве.

Преимущества:

- Хорошо справляется с нелинейными данными.
- Сохраняет локальные структуры.

Недостатки:

- Чувствителен к шуму.
- Требует тонкой настройки параметров.

```python
from sklearn.datasets import load_digits
from sklearn.manifold import LocallyLinearEmbedding
import matplotlib.pyplot as plt
import numpy as np

# Загрузка набора данных digits
digits = load_digits()
X = digits.data
y = digits.target

# Создание объекта LLE
lle = LocallyLinearEmbedding(
    n_neighbors=10,      # Количество ближайших соседей
    n_components=2,      # Размерность выходного пространства
    random_state=42      # Для воспроизводимости результатов
)

# Применение преобразования к первым 100 точкам для наглядности
X_reduced = lle.fit_transform(X[:100])

# Визуализация результата
plt.figure(figsize=(10, 8))
scatter = plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=y[:100], cmap='viridis')
plt.colorbar(scatter)
plt.title('Результат уменьшения размерности методом LLE')
plt.xlabel('Первая компонента')
plt.ylabel('Вторая компонента')

# Сохранение графика
plt.savefig('lle_result.png')
```
